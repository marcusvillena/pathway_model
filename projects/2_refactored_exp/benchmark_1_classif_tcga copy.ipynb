```python
def run_expt(counts_name:str, counts_data:d.TCGA, num_trials:int=5):
    # data
    kegg = d.KEGG(
        relation_filepath=dataset_dir/'other'/'relation_ohe.csv', 
        counts_data=counts_data,
    )
    dataset = d.GraphDataset(counts_data, kegg, kegg)

    # models
    glm = MultiLatentModel(
        dataset=dataset,
        embed_dim = 128,
        method='node', # !!!

        norm_class = LogCounts,
        encoder_class = nn.Linear,
        pooling_class = AttentionSetPooling,
        mlp = False,
        variational = False, # !!!

        hidden_dims = 0, # !!!
        act_fn = nn.ReLU, 
        norm_fn = 'layer', 
        end_fn = False,

        norm_kwargs = {'libnorm':True, 'znorm':True, 'learnable':True,}
    )

    vae = MultiLatentModel(
        dataset=dataset,
        embed_dim = 128,
        method='node', # !!!

        norm_class = LogCounts,
        encoder_class = nn.Linear,
        pooling_class = AttentionSetPooling,
        mlp = False,
        variational = False, # !!!

        hidden_dims = 1, # !!!
        act_fn = nn.ReLU, 
        norm_fn = 'layer', 
        end_fn = False,

        norm_kwargs = {'libnorm':True, 'znorm':True, 'learnable':True,}
    )

    pvae = MultiLatentModel(
        dataset=dataset,
        embed_dim = 128,
        method='set', # !!!


        norm_class = LogCounts,
        encoder_class = nn.Linear,
        pooling_class = AttentionSetPooling,
        mlp = False,
        variational = False, # !!!

        hidden_dims = 1, # !!!
        act_fn = nn.ReLU, 
        norm_fn = 'layer', 
        end_fn = False,

        norm_kwargs = {'libnorm':True, 'znorm':True, 'learnable':True,}
    )

    # trainer
    trainer = ClassifTrainer(
        lr=1e-3, 

        early_stop=True,
        stop_metric='accuracy',
        stop_kwargs={'mode':'max'},

        kw_keys={'input':'y_logits', 'target':'y'},
        metric_keys={'pred':'input', 'target':'target'},
        loss_class=nn.CrossEntropyLoss,
    )
    
    # expt
    expt = Experiment(
        num_trials=num_trials, # scVI: 5 trials
        num_epochs=300, # scVI: 200 epochs
        dataset=dataset,
        device=device,
        batch_size=128
    )

    expt.add_config(
        name='gene-linear',
        model=glm,
        trainer=trainer,
    )

    expt.add_config(
        name='gene-vae',
        model=vae,
        trainer=trainer,
    )

    expt.add_config(
        name='pathway-vae',
        model=pvae,
        trainer=trainer,
    )

    expt.run_experiment(
        f'benchmark_1_classif_tcga_{counts_name}',
        report_metrics=['loss','kld','nb','mae', 'accuracy', 'precision', 'f1'],
        save_model=True,
        save_csv=True,
        save_params=True,
        save_values=True,
        verbose=False,
    )
```

```python
for k,v in tcga.items():
    run_expt(k, v, num_trials=25)

# for k,v in tcga.items():
#     run_expt(k, v, num_trials=20)

```
